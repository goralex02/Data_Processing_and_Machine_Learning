Вначале прочитаем данные.

```python
import pandas as pd
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

Посмотрим на числовые данные.

```python
train.describe()
```

Столбец id не нужен нам при обучении, но пригодится при записи итога.

```python
train = train.drop(columns = ['id'])
id_column = test.pop('id')
```

Теперь посмотрим на нечисловые данные.

```python
train.describe(include=object)
```

Преобразуем Total_Payment в числовые данные, одновременно превратив пустые строки в ноль.

```python
test['Total_Payment'] = test['Total_Payment'].replace(' ', '0.0')
train['Total_Payment'] = train['Total_Payment'].replace(' ', '0.0')
```

```python
test['Total_Payment'] = pd.to_numeric(test['Total_Payment'])
train['Total_Payment'] = pd.to_numeric(train['Total_Payment'])
```

Используем LabelEncoder, чтобы превратить категориальные данные в числовые.

```python
from sklearn.preprocessing import LabelEncoder

for column in train.columns:
    if train[column].dtype == object:
        le = LabelEncoder().fit(list(train[column].values) + list(test[column].values))
        train[column] = le.transform(train[column].values)
        test[column] = le.transform(test[column].values)
```

Посмотрим на получившиеся данные.

```python
train.describe()
```

Разделим данные на тренировочную и тестовую часть.

```python
from sklearn.model_selection import train_test_split

X = train.drop(columns=['Leave_Next_Month'])
y = train['Leave_Next_Month']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=777)
```

Используем xgboost, взяв его из примера.

```python
import xgboost as xgb

params_xgb = {
    "objective": "binary:logistic",
    "eval_metric": 'logloss',
    "eta": 0.05,
    "max_depth": 2,
    "subsample": 0.8,
    "colsample_bytree": 0.8,
    "verbosity": 0,
    "silent": True,
}
num_boost_round = 70
early_stopping_rounds = 10
```

Изменим параметры модели для повышения точности.

```python
import xgboost as xgb

params_xgb = {
    "objective": "binary:logistic",
    "eval_metric": 'logloss',
    "eta": 0.047,
    "max_depth": 2,
    "subsample": 0.78,
    "colsample_bytree": 0.78,
    "verbosity": 0,
    "silent": True,
}
num_boost_round = 120
early_stopping_rounds = 15
```

Обучим модель.

```python
dtrain = xgb.DMatrix(X_train.values, y_train.values)
dvalid = xgb.DMatrix(X_test.values, y_test.values)
watchlist = [(dtrain, 'train'), (dvalid, 'eval')]
gbm = xgb.train(params_xgb, dtrain, num_boost_round, evals=watchlist,
                early_stopping_rounds=early_stopping_rounds, verbose_eval=10)
```

Выведем метрики.

```python
import numpy as np
from sklearn.metrics import accuracy_score, roc_auc_score

pred = gbm.predict(xgb.DMatrix(X_test.values), ntree_limit=gbm.best_iteration + 1)
accuracy = accuracy_score(y_test.values, np.round(pred))
auc = roc_auc_score(y_test.values, pred)
print('Accuracy: {:.2f} %, ROC AUC: {:.2f}'.format(100*accuracy, auc))
```

Сделаем предсказания на данных из файла test.csv (тут пригодился столбец id, ранее изъятый из данных).

```python
pred = gbm.predict(xgb.DMatrix(test.values), ntree_limit=gbm.best_iteration + 1)

result= {'id' : id_column.values,
         'Leave_Next_Month': pred}
```

Запишем результат в .csv файл.

```python
pd.DataFrame(result).to_csv('result.csv', index=False)
```
